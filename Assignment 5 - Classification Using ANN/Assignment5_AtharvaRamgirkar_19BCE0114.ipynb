### 1. Importing Libraries

```python
import pandas as pd
import numpy as np

# To Initialize the model
from tensorflow.keras.models import Sequential 
# To Add layers
from tensorflow.keras.layers import Dense 
# To Add Droupout Layers
from tensorflow.keras.layers import Dropout

import tensorflow as tf
import keras

# Split the Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
import matplotlib.pyplot as plt
%matplotlib inline
```

### 4. Removing Unnecessary Columns

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
df.head()
```

```python
df['deposit'] = np.where(df['deposit']=="yes",1,0)
```

### 5. Spliting the Dataset into Target and Features

```python
X = df.iloc[:,0:48].values
y = df.iloc[:,48].values
```

```python
X
```

```python
y
```

### 6. Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 test_size=0.3,
                                                 random_state=0,
                                                 stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 7. Model Building

#### 7.1 Initialize the Model

```python
# Feed Forward 
model = Sequential()
```

#### 7.2 Add Input Layer 

```python
# Input Layer
model.add(Dense(units=48,
                kernel_initializer="random_uniform",
                activation="relu"))
```

#### 7.3 Add Hidden Layers and Dropout Layers

```python
# hidden layer
model.add(Dense(units=48,
                kernel_initializer="random_uniform",
                activation="relu"))

# Dropout Layer 1
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(units=96,
                kernel_initializer="random_uniform",
                activation="relu"))

# Dropout Layer 2
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(units=48,
                kernel_initializer="random_uniform",
                activation="relu"))
```

#### 7.4 Add Output Layer

```python
# output layer
model.add(Dense(units=1,
                kernel_initializer="random_uniform",
                activation="sigmoid"))
```

### 8. Compile Model

```python
model.compile(optimizer="sgd",
                  loss="binary_crossentropy",
                  metrics=["accuracy"])
```

### 9. Model Testing

```python
history = model.fit(X_train,y_train,epochs=1000,validation_data=(X_test,y_test))
```

### 10. Final Accuracy

Train Accuracy      : **98.89**  
Validation Accuracy : **81.13**  
Vaidation Loss      : **1.7249**
