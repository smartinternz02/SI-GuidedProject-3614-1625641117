### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Getting the data

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

### Matplotlib Visualizations

#### Line Plot

```python
plt.plot(np.linspace(10,20),
         np.linspace(15,25),
         color = "blue",
         label = "x v/s y")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

#### Scatter Plot

```python
plt.scatter(iris.sepal_length,iris.sepal_width)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Width v/s Length")
plt.show()
```

#### Bar Plot

```python
plt.bar(iris.species,iris.petal_length,color = "blue")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.title("Petal Length DIstribution of different Iris Species")
plt.show()
```

#### Pie Chart

```python
plt.pie(list(iris.species.value_counts()),
        labels = iris.species.unique(),
        autopct= '%0.3f%%',
        colors = ["purple","pink","yellow"])
plt.show()
```

### Seaborn Visualizations

#### Scatter Plot

```python
sns.scatterplot(x="petal_length",y = "petal_width",data = iris)
```

#### Line Plot

```python
sns.lineplot(x = "petal_length",y = "petal_width",data = iris)
```

```python
sns.lineplot(x = "sepal_length",y = "sepal_width",data = iris)
```

#### Count Plot

```python
sns.countplot(x = "species",data = iris)
```

#### Box Plot

```python
sns.boxplot(x ="petal_width",data = iris)
```

```python
sns.boxplot(x ="petal_length",data = iris)
```

```python
sns.boxplot(x ="sepal_length",data = iris)
```

#### Categorical Plot

```python
iris["petal_length_cat"] = np.where(iris["petal_length"]<2,
                                    "<2",
                                    np.where(iris["petal_length"]<4,
                                             "<4",
                                             ">4"))
```

```python
sns.catplot(x="petal_length_cat",
            y= "sepal_length",
            hue = "species",
            kind = "bar",
            data = iris)
```

#### Pair Plot

```python
sns.pairplot(iris)
```

#### Correlation Matrix

```python
iris.corr()
```

#### Heat Map

```python
sns.heatmap(iris.corr(),
           annot=True,
           fmt=".2f", 
           linewidths=5,
           cmap=sns.diverging_palette(0,
                                      230,
                                      90,
                                      60,
                                      as_cmap=True),
           vmin=-1,
           vmax=1, 
           square=True)
```
